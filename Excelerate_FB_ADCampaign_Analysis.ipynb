```python
def Plot(df):
    ix = 1
    fig = plt.figure(figsize = (12,8))
    for c in list(df.columns):
        if df[c].nunique() >= 20:
            if ix <= 3:
                ax2 = fig.add_subplot(2,3,ix+3)
                ax2.tick_params(axis='x', labelrotation=90)
        
                sns.boxplot(y=df[c], ax=ax2) #boxplot for outliers     
                fig.tight_layout()
            ix = ix +1
            if ix == 4: 
                fig = plt.figure(figsize = (12,8))
                ix = 1
```

```python
Plot(df)
```

## Bivariate Analysis

```python
plt.figure(figsize=(20,10))

plt.subplot(2, 3, 1)
plt.bar(df['Campaign ID'], df['Reach'])
plt.title('Campaign ID vs Reach')
plt.xlabel('Campaign ID')
plt.ylabel('Reach')
plt.xticks(rotation=90)

plt.subplot(2, 3, 2)
plt.bar(df['Campaign ID'], df['Impressions'])
plt.title('Campaign ID vs Impressions')
plt.xlabel('Campaign ID')
plt.ylabel('Impressions')
plt.xticks(rotation=90)

plt.show()
```

```python
plt.figure(figsize=(20,10))

plt.subplot(2, 3, 1)
plt.bar(df['Campaign ID'], df['Clicks'])
plt.title('Campaign ID vs Clicks')
plt.xlabel('Campaign ID')
plt.ylabel('Clicks')
plt.xticks(rotation=90)

plt.subplot(2, 3, 2)
plt.bar(df['Campaign ID'], df['Unique Clicks'])
plt.title('Campaign ID vs Unique Clicks')
plt.xlabel('Campaign ID')
plt.ylabel('Unique Clicks')
plt.xticks(rotation=90)

plt.subplot(2, 3, 3)
plt.bar(df['Campaign ID'], df['Unique Link Clicks (ULC)'])
plt.title('Campaign ID vs Unique Link Clicks')
plt.xlabel('Campaign ID')
plt.ylabel('Unique Link Clicks')
plt.xticks(rotation=90)

plt.show()
```

```python
plt.figure(figsize=(20,10))

plt.subplot(2, 3, 1)
plt.bar(df['Campaign ID'], df['Click-Through Rate (CTR)'])
plt.title('Campaign ID vs Click-Through Rate')
plt.xlabel('Campaign ID')
plt.ylabel('Click-Through Rate')
plt.xticks(rotation=90)

plt.subplot(2, 3, 2)
plt.bar(df['Campaign ID'], df['Unique Click-Through Rate (Unique CTR)'])
plt.title('Campaign ID vs Unique Click-Through Rate')
plt.xlabel('Campaign ID')
plt.ylabel('Unique Click-Through Rate')
plt.xticks(rotation=90)

plt.show()
```

```python
plt.figure(figsize=(20,10))

plt.subplot(2, 3, 1)
plt.bar(df['Campaign ID'], df['Cost Per Click (CPC)'])
plt.title('Campaign ID vs Cost Per Click')
plt.xlabel('Campaign ID')
plt.ylabel('Cost Per Click')
plt.xticks(rotation=90)

plt.subplot(2, 3, 2)
plt.bar(df['Campaign ID'], df['Cost per Result (CPR)'])
plt.title('Campaign ID vs Cost per Result')
plt.xlabel('Campaign ID')
plt.ylabel('Cost per Result')
plt.xticks(rotation=90)

plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.bar(df['Campaign ID'], df['Amount Spent in INR'])
plt.title('Campaign ID vs Amount Spent in INR')
plt.xlabel('Campaign ID')
plt.ylabel('Amount Spent in INR')
plt.xticks(rotation=90)
plt.show()
```
